## Numbers and Strings
Well good job for making that may changes the world (Applause!!!)
Sorry, but we are not there yet. Python interpreter allows you to perform more operations on your end and on the fly like addition, substraction, multiplication, division, so on and so forth.
That was for number, however for strings you can do some concatenation, slicing, calculate the length of a string, indexing, etc.
Nice han ! Let's see them

````python
############ Numbers
# Addition
10 + 3
# Substraction
8 - 420
# Division
10 / 5
# Multiplication
2 * 3
# Raising to power
3 ** 2
# Floor division
10 // 3
########### Strings
'py' 'thon'
'python ' + 'is created by ' + 'Guido Rossum'
'I\'m a little bit skeptical.'
"I was not, don't worry"
full_name = "Ousmane Faye"
# Split by space and seize the first name which has index 0
first_name = full_name.split(" ")[0]
# then seize the last name which follows with index 1
last_name = full_name.split(" ")[1]
# Print them on your screen
print(first_name," ", last_name)
# Seize any in the full_name
first_word = full_name[0]
last_word = full_name[-1]
rand_word = full_name[3]
print("First word: ", first_word)
print("Last word: ", last_word)
print("Random word: ", rand_word)
print("Length of full name is: ", len(full_name))
````
